```python
file_path = r'C:\Users\jeffr\Python\reading_data\beer.txt'
def open_file(fi_path, seperator = ',', head = 0, encode = 'ISO-8859-1'):
  """
  This function reads csv and excel files. the parameter seperator has a default
  value of ',' , however the user can input any other seperator. The header
  default is 0 which means the headers are in row 0 of the dataset
  """
  try:
    filename, extension = fi_path.split('.')[-2:]
    if extension == 'xlsx':
      file_data = pd.read_excel(fi_path)
    else:
      file_data = pd.read_csv(fi_path, sep = seperator, header = head, encoding = encode)
    return file_data

  except Exception as err:
    print(f'Error reading file: {err}')

open_file(file_path, seperator = ' ')
```
